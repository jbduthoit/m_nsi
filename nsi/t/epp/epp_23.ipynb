 ## Exo 2

La fonction fusion prend deux listes L1,L2 d’entiers triées par ordre croissant et les 
fusionne en une liste triée L12 qu’elle renvoie. 
Compléter le code suivant :

```python
def fusion(L1,L2): 
    n1 = len(L1) 
    n2 = len(L2) 
    L12 = [0]*(n1+n2) 
    i1 = 0 
    i2 = 0 
    i = 0 
    while i1 < n1 and ... : 
        if L1[i1] < L2[i2]: 
            L12[i] = ... 
            i1 = ... 
        else: 
            L12[i] = L2[i2]  
            i2 = ... 
        i += 1 
    while i1 < n1: 
     L12[i] = ... 
     i1 = i1 + 1 
     i = ... 
    while i2 < n2: 
     L12[i] = ... 
     i2 = i2 + 1 
     i = ... 
    return L12
```

```python
assert fusion([1,6,10],[0,7,8,9]) == [0, 1, 6, 7, 8, 9, 10]
```
